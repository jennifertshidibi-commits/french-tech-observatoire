```python
import pyarrow.parquet as pq

fichier = "../data/stock-stocketablissement-parquet.parquet"
parquet_file = pq.ParquetFile(fichier)

print("Nombre de lignes:", parquet_file.metadata.num_rows)
print("Nombre de colonnes:", parquet_file.metadata.num_columns)
```

```python
# je liste toutes les colonnes du fichier avec une boucle for 

colonnes = parquet_file.schema.names
print(f"Total: {len(colonnes)} colonnes\n")
for i, col in enumerate(colonnes):
    print(f"{i+1:2d}. {col}")
```
